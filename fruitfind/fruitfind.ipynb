# The network

As network we take a convolutional neural network CNN.

## Convolution 

_Convolutional_ means we have a so-called _kernel_ that sweep over an input matrix to produce and output matrix. A kernel is a (small) matrix with _weights_. Typically the weights sum to 1. The image below shows a 3Ã—3 kernel.

![convolution kernel.png](convolution-kernel.png)

A kernel sweeps over an input matrix. The diagram below shows the kernel on an arbitrary locations ("x"). When the kernel is at location "x" it multiplies the 9 cells around (and including) "x" with the kernel weights, sums all 9 values, and this weighted average is output. 

![convolution image.png](convolution-img.png)

The kernel sweep over all locations, and computes the weighted average of all cells. The diagram below shows an example with an input matrix of 5Ã—4 where the 3Ã—3 kernel from above is used to compute an output matrix. For example, (0.1Ã—100 + 0.2Ã—110 + 0Ã—120) + (0.2Ã—100 + 0.3Ã—120 + 0.1Ã—140) + (0Ã—100 + 0.1Ã—130 + 0Ã—160) = 115. The behavior at the borders is special, easiest is to decrease the output size with the kernel size (minus 1).

![convolution numerical.png](convolution-num.png)



## Neurons

In a convolutional neural network (CNN), we have a kernel that sweeps over an "image" at layer _n_ to produce and "image" at layer _n+1_. The weights of the kernel is determined by learning. This means that the amount of weights is relatively small: we have to map a (large) input image to a(large) output image, but we only have to train the weights in a small kernel.

There is one complication, the "image" at each layer has multiple planes. This means that the kernel is not two-dimensional (2D) but a three-dimensional (3D) matrix. In the diagram below the input "image" has a width of 8, a height of 7 and a depth of 2 (light blue). This means that a kernel of 3Ã—3 is actually a matrix of 3Ã—3Ã—2 (dark blue). The "neuron" (see the creditcard notebook for an introduction) has 18 weights, and on top of that a bias.

![Weights](weights.png)

With these 19 parameters, the dark green cell can be computed. By sweeping the kernel, one complete plane in the output "image" can be computed. If there are 4 output planes, we need 4 of those kernels, and thus 4Ã—19 or 76 parameters need to be trained.

In the world of CNNs, the planes are often refered to as _features_.

## Funnel

CNN are well suited to do image classification. We are going to construct a 10 layer convolutional neural network. It has a typical funnel shape. We begin with the input layer, which has an input resolution of 100Ã—100 and a depth of 3 planes (the three color channels: red, green and blue). When we go through the layers the resolution becomes smaller and smaller, but the number of planes ("features") increases. At the first step, we go from 100Ã—100Ã—3 to 98Ã—98Ã—16, the increase in depth is a choice from us, the decrease in resolution is the border effect of convolution.

![CNN Funnel](funnel.png)

At the third step we not only double the number of planes (the depth) but we also make an explicit jump in resolution by clubbing 2Ã—2 pixels. So we go from 96Ã—96Ã—16 to 47Ã—47Ã—32.

## The code

let us now setup the network with _keras_. We have a sequential network of convolution layers (`Conv2D`). For each layer we use a `(3,3)` kernel, and `'relu'` as activation function. The `padding='valid'` means that we chose to shrink the border with the kernel size. We have the resolution (`strides=(2,2)`) now and then. At the end we have a layer with a resolution of 7Ã—7 and a feature depth of 128. We need 131, so we append a `Conv2D(131,(1,1))`. Finally we average the 7Ã—7 image to one pixel (`GlobalAveragePooling2D`).

````python
from tensorflow import keras
model = keras.models.Sequential()
# 100Ã—100Ã—3
model.add(keras.layers.Conv2D( 16,(3,3),padding='valid',activation='relu',input_shape=(100,100,3)))
# 98Ã—98Ã—16
model.add(keras.layers.Conv2D( 16,(3,3),padding='valid',activation='relu'))
# 96Ã—96Ã—16
model.add(keras.layers.Conv2D( 32,(3,3),padding='valid',activation='relu',strides=(2,2)))
# 47Ã—47Ã—32
model.add(keras.layers.Conv2D( 32,(3,3),padding='valid',activation='relu'))
# 45Ã—45Ã—32
model.add(keras.layers.Conv2D( 64,(3,3),padding='valid',activation='relu',strides=(2,2)))
# 22Ã—22Ã—64
model.add(keras.layers.Conv2D( 64,(3,3),padding='valid',activation='relu'))
# 20Ã—20Ã—64
model.add(keras.layers.Conv2D(128,(3,3),padding='valid',activation='relu',strides=(2,2)))
# 9Ã—9Ã—128
model.add(keras.layers.Conv2D(128,(3,3),padding='valid',activation='relu'))
# 7Ã—7Ã—128
model.add(keras.layers.Conv2D(131,(1,1)))
# 7Ã—7Ã—131
model.add(keras.layers.GlobalAveragePooling2D())
# 1Ã—1Ã—131
model.add(keras.layers.Activation('softmax'))
````

Let us now compile the network and check the parameters.

````python
model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
model.summary()
````

The first step brings us from 100Ã—100Ã—3 to 98Ã—98Ã—16 with a 3Ã—3 kernel.

Recall that the kernel has a depth of 3 and 1 bias, so we need 3Ã—3Ã—3+1 or 28 parameters per output plane, and we have 16 of those, so in total 16Ã—28 = 448. This matches the first row in the table above.

Let's do one more. Step 3 goes from 96Ã—96Ã—16 to 47Ã—47Ã—32, again with a 3Ã—3 kernel. The kernel now has a depth of 16, so the we need 3Ã—3Ã—16+1 or 145 parameters per output pane. We have 32 output panes so 32Ã—145 or 4640 in total. This matches the third row.
